```python
import pandas as pd

# List of desired columns
columns_to_keep = [
    'full_date', 'full_date_year_num', 'month_num', 'month_day_num', 
    'quarter_num', 'Brand', 'divison', 'category', 'Product', 
    'marketplace', 'daily_revenue_usd', 'daily_units_sold', 'ppc_sales', 
    'ppc_cost', 'orders', 'cogs', 'fbafees', 'promo_amount', 
    'contribution_margin', 'referralfees', 'fba_fulfillment_fees', 
    'promo_units', 'shipped', 'ret_date_return_amount_usd', 
    'ret_date_return_promo_amount_usd', 'ret_date_return_quantity', 
    'returned_units', 'return_usd', 'ops_net_tax'
]

# Create a new DataFrame with only the specified columns
retail_filtered =retail[columns_to_keep]


```

```python
import pandas as pd
import numpy as np

# List of columns with blanks
columns_with_blanks = [
    'daily_revenue_usd', 'daily_units_sold', 'ppc_sales', 'ppc_cost', 'orders',
    'cogs', 'fbafees', 'promo_amount', 'contribution_margin', 'referralfees',
    'fba_fulfillment_fees', 'promo_units', 'shipped', 'ret_date_return_amount_usd',
    'ret_date_return_promo_amount_usd', 'ret_date_return_quantity', 'returned_units',
    'return_usd', 'ops_net_tax'
]

# Create a flag column to indicate rows that had blanks(no values)(1 is True,, 0 is False)
retail_filtered['had_blanks'] = retail_filtered[columns_with_blanks].isnull().any(axis=1).astype(int)

# Fill blanks with zeros to ensure data integrity
retail_filtered[columns_with_blanks] = retail_filtered[columns_with_blanks].fillna(0)

# Print summary to verify
#print(f"Rows with originally blank values: {retail_filtered['had_blanks'].sum()}")
#print(f"Percentage of rows with filled blanks: {retail_filtered['had_blanks'].mean()*100:.2f}%")
```
